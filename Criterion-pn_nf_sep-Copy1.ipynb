```python
from scipy.integrate import quad

N=100
pn_max_mi=100000
pn_summi=100000
pn_deltami=100000
nf_max_mi=100000
nf_summi=100000
nf_deltami=100000

def integrand(r, a0, a1,a2):
    return 2.0*pi* a0*r / ((1.0 + a1*r + a2*r*r)*(1.0 + 0.0080*r))

for r1 in range(60,160,10):
    print(r1)
    for r2 in range(r1+40,230,10):
        aver=np.zeros(8)
        print(r2)
        s1=0
        s2=0
        cri=[]
        for i in range(1,9):
            L=len(df['p0'][i])
            cri_j=np.zeros(L)
            for j in range(L):
                a0 = df['p0'][i,j]
                a1 = df['p1'][i,j]
                a2 = df['p2'][i,j]
                s1 = quad(integrand, 0, r1, args=(a0,a1,a2))
                s2 = quad(integrand, r1, r2, args=(a0,a1,a2))
                cri_j[j]=s1[0]/s2[0]
                aver[i-1]+=cri_j[j]/L                
            cri.append(cri_j)
        bpn_min=min(aver[1],aver[5])
        bpn_max=max(aver[0],aver[4])
        bpn=bpn_min
        
#-------------------------------p-N-----------------------------------------

        for bpn in np.arange(bpn_min,bpn_max,1/N):      #Сразу считаю ошибки.
            pn_p=0
            pn_n=0
            pn_p2=0
            pn_n2=0
            for j in np.arange(1000):
                if j<len(df['p0'][1]) and cri[0][j]<bpn: pn_p+=1/len(df['p0'][1])
                if j<len(df['p0'][2]) and cri[1][j]>bpn:  pn_n+=1/len(df['p0'][2])                    
                if j<len(df['p0'][5]) and cri[4][j]<bpn:  pn_p2+=1/len(df['p0'][5])
                if j<len(df['p0'][6]) and cri[5][j]>bpn:  pn_n2+=1/len(df['p0'][6])
            
            # 1 способ поиска min ошибки. Min(max)
            pn_max=max(pn_p,pn_n,pn_p2,pn_n2)
            if pn_max<pn_max_mi:
                pn_max_mi=pn_max
                pn_border_max=bpn
                pn_r1_max=r1
                pn_r2_max=r2
            
            #2 способ.                 
            delta = abs(pn_p-pn_p2)+abs(pn_n-pn_n2)   
            if delta<pn_deltami:
                pn_deltami = delta
                pn_delta=pn_max
                pn_border_delta=bpn
                pn_r1_delta=r1
                pn_r2_delta=r2
           
            #3 способ поиска min. Две максимальные ошибки наиболее близки друг к другу. Это предполо
            #жено из того. Если посмотреть на файл с ошибками для разных границ, то можно заметить что 
            #ошибки в одном случае растут, в другом убывают. Из этого я сделала вывод, что можно не просто 
            #искать  max(min), а наиболее близкие друг к другу две наивысшие ошибки
            if pn_p<=pn_n: pn_p=pn_n
            if pn_p2<=pn_n2: pn_p2=pn_n2
            summ=(pn_p+pn_p2)/2
            summa=abs(pn_p-summ)+abs(pn_p2-summ)            
            if summa<pn_summi:
                pn_summi=summa
                pn_border_sum=bpn
                pn_sum=max(pn_p,pn_p2)
                pn_r1_sum=r1
                pn_r2_sum=r2
        bnf_min=min(aver[2],aver[6])
        bnf_max=max(aver[1],aver[5])
        bnf=bnf_min
#--------------------------------N-Fe-----------------------------------------

        for bnf in np.arange(bnf_min,bnf_max,1/N):      #Сразу считаю ошибки.
            nf_n=0
            nf_f=0
            nf_n2=0
            nf_f2=0
            for j in np.arange(1000):
                if j<len(df['p0'][2]) and cri[1][j]<bnf: nf_n+=1/len(df['p0'][2])
                if j<len(df['p0'][3]) and cri[2][j]>bnf:  nf_f+=1/len(df['p0'][3])                    
                if j<len(df['p0'][6]) and cri[5][j]<bnf:  nf_n2+=1/len(df['p0'][6])
                if j<len(df['p0'][7]) and cri[6][j]>bnf:  nf_f2+=1/len(df['p0'][7])

            # 1 способ поиска min ошибки. Min(max)
            nf_max=max(nf_n,nf_f,nf_n2,nf_f2)
            if nf_max<nf_max_mi:
                nf_max_mi=nf_max
                nf_border_max=bnf
                nf_r1_max,nf_r2_max=r1,r2
            
            #2 способ.                 
            delta = abs(nf_n-nf_n2)+abs(nf_f-nf_f2)   
            if delta<nf_deltami:
                nf_deltami = delta
                nf_delta=nf_max
                nf_border_delta=bnf
                nf_r1_delta,nf_r2_delta=r1,r2
           
            #3 способ поиска min. Две максимальные ошибки наиболее близки друг к другу. Это предполо
            #жено из того. Если посмотреть на файл с ошибками для разных границ, то можно заметить что 
            #ошибки в одном случае растут, в другом убывают. Из этого я сделала вывод, что можно не просто 
            #искать  max(min), а наиболее близкие друг к другу две наивысшие ошибки
            if nf_n<=nf_f: nf_n=nf_f
            if nf_n2<=nf_f2: nf_n2=nf_f2
            summ=(nf_n+nf_n2)/2
            summa=abs(nf_n-summ)+abs(nf_n2-summ)            
            if summa<nf_summi:
                nf_summi=summa
                nf_border_sum=bnf
                nf_sum=max(nf_n,nf_n2)
                nf_r1_sum,nf_r2_sum=r1,r2
                
print(pn_max_mi,pn_border_max,pn_r1_max,pn_r2_max)
print(pn_delta,pn_border_delta,pn_r1_delta,pn_r2_delta)
print(pn_sum,pn_border_sum,pn_r1_sum,pn_r2_sum)

print(nf_max_mi,nf_border_max,nf_r1_max,nf_r2_max)
print(nf_delta,nf_border_delta,nf_r1_delta,nf_r2_delta)
print(nf_sum,nf_border_sum,nf_r1_sum,nf_r2_sum)
```
