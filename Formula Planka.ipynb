```python
import matplotlib.pyplot as plt
import numpy as np

herg=6.63*10**(-27)  #erg*s
hev=4.14*10**(-15)   #eV*s
c=3*10**(10)         # cm/s
k=8.62*10**(-5)   #eV/K
T=10  #K


for i in range(3):
    T1=0.9*T   #cian
    T2=T
    T3=1.1*T  #red

nu = np.arange(10**8, 0.5*10**13, 10**10)
B1 = 2*nu*herg*nu*nu/(c*c*(np.exp(hev*nu/(k*T1))-1))
B2 = 2*nu*herg*nu*nu/(c*c*(np.exp(hev*nu/(k*T2))-1))
B3 = 2*nu*herg*nu*nu/(c*c*(np.exp(hev*nu/(k*T3))-1))
#print(nu*herg*nu*nu,c*c*(np.exp(hev*nu/(k*T))-1))
#l = np.arange(10**(-6), 10**(-4), 10**(-6))
#Bm=2*herg*c**2/(l**5*(np.exp(hev*c/(l*k*T))-1))

fig= plt.subplot()

plt.plot(nu, B1, color='c')
plt.plot(nu, B2, color='y')
plt.plot(nu, B3, color='r')
plt.xlabel('Hz')
plt.ylabel('B, эрг/(см^2*с*Гц)')
plt.title('Формула Планка для T=10')

plt.grid(True, linestyle='-.')

plt.show()
```
